# Problemas de distribución de probabilidad

## Distribución Binomial:

En teoría de la probabilidad y estadística, la distribución binomial es la distribución de probabilidad discreta que da sólo dos resultados posibles en un experimento; por ejemplo, al lanzar una moneda, puede salir cara o cruz.

```python
# - Importe de las librerías a emplear:

from scipy.stats import binom
import numpy as np
import matplotlib.pyplot as plt

# - Definición de todas las funciones relacionadas a la Distribución Binomial:

def dbinom(x, size, prob = 0.5):
    
    # - Calcula la estimación puntual de la distribución binomial.
    
    result = binom.pmf(k = x, n = size, p = prob, loc = 0)
    return result

def pbinom(q, size, prob = 0.5):

    # - Calcula el acumulado de la distribución binomial.

    result = binom.cdf(k = q, n = size, p = prob, loc = 0)
    return result

def qbinom(p, size, prob = 0.5):

    # - Calcula la función cuantil a partir de la distribución binomial.

    result = binom.ppf(q = p, n = size, p = prob, loc = 0)
    return result

def rbinom(n, size, prob = 0.5):

    # - Genera variables aleatorias a partir de la distribución binomial. 

    result = binom.rvs(n = size, p = prob, size = n)
    return result
```

```python
# - Los datos a emplear para la distribución binomial son: El número de ensayos siendo n=10, la probabilidad de éxito siendo p=0.2 o 20%.

np.random.seed(42)

print(f"Probabilidad de que una distribución binomial (10, 0.2) tome el valor de 2: {round(dbinom(2, size = 10, prob = 0.2) * 100, 2)}%")
print()
print(f"Probabilidad de que una distribución binomial (10, 0.2) tome un valor menor que 2: {round(pbinom(2, size = 10, prob = 0.2) * 100, 2)}%")
print()
print(f"Valor que de una distribución binomial (10, 0.2) tiene una probabilidad acumulada de 0.9: {round(qbinom(0.9, size = 10, prob = 0.2), 2)}")
print()
print(f"Generación de 2000 valores aleatorios de una distribución binomial (10, 0.2): {rbinom(2000, size = 10, prob = 0.2)}")
```

```python
np.random.seed(42)

# - Declaración de variables:

size = 10
prob = 0.2
data = rbinom(2000, size = 10, prob = 0.2)

pmf = dbinom(data, size = size, prob = prob)
cdf = pbinom(data, size = size, prob = prob)
ppf = qbinom(data, size = size, prob = prob)

# - Distribución de los gráficos:

fig, axis = plt.subplots(1, 3, figsize = (20, 5))

# - Parámetros de los gráficos:

axis[0].stem(data, pmf, 'r-')
axis[0].set_xlabel("Número De éxitos", fontsize=12)
axis[0].set_ylabel("Probabilidad", fontsize=12)
axis[0].grid(True)

axis[1].stem(data, cdf, 'g-')
axis[1].set_xlabel("Número De éxitos", fontsize=12)
axis[1].set_ylabel("Probabilidad Acumulada", fontsize=12)
axis[1].grid(True)

axis[2].stem(data, ppf, 'b-')
axis[2].set_xlabel("Percentil", fontsize=12)
axis[2].set_ylabel("Valor De La Variable Aleatoria", fontsize=12)
axis[2].grid(True)

# - Visualización:

plt.tight_layout()
plt.show()
```

## Distribución Normal:

```python
# - Importe de las librerías a emplear:

from scipy.stats import norm

# - Definición de todas las funciones relacionadas a la Distribución Normal:

def dnorm(x, mean = 0, std = 1):
    
    # - Calcula la estimación puntual de la distribución normal.
    
    result = norm.pdf(x, loc = mean, scale = std)
    return result

def pnorm(q, mean = 0, std = 1):
    
    # - Calcula el acumulado de la distribución normal.
    
    result = norm.cdf(q, loc = mean, scale = std)
    return result

def qnorm(p, mean = 0, std = 1):
    
    # - Calcula la función cuantil a partir de la distribución normal.
    
    result = norm.ppf(p, loc = mean, scale = std)
    return result

def rnorm(n, mean = 0, std = 1):
    
    # - Genera variables aleatorias a partir de la distribución normal.
    
    result = norm.rvs(loc = mean, scale = std, size = n)
    return result
```

```python
# - Los datos a emplear para la distribución normal son: Media siendo mean=0, Desviación estándar siendo std=1.

np.random.seed(42)

print(f"Probabilidad de que una distribución normal (0, 1) tome el valor de 0: {round(dnorm(0, mean = 0, std = 1) * 100, 2)}%")
print()
print(f"Probabilidad de que una distribución normal (0, 1) tome un valor menor que 0: {round(pnorm(0, mean = 0, std = 1) * 100, 2)}%")
print()
print(f"Valor que de una distribución normal (0, 1) tiene una probabilidad acumulada de 0.9: {round(qnorm(0.9, mean = 0, std = 1), 2)}")
print()
print(f"Generación de 2000 valores aleatorios de una distribución normal (0, 1): {rnorm(2000, mean = 0, std = 1)}")
```

```python
np.random.seed(42)

# - Declaración de variables:

mean = 0
std = 1
data = np.arange(-5, 5, 0.01)

pmf = dnorm(data, mean = mean, std = std)
cdf = pnorm(data, mean = mean, std = std)
ppf = qnorm(data, mean = mean, std = std)

# - Distribución de los gráficos:

fig, axis = plt.subplots(1, 3, figsize = (20, 5))

# - Parámetros de los gráficos:

axis[0].plot(data, pmf, "r-")
axis[0].set_xlabel("Valor de la Variable Aleatoria", fontsize=12)
axis[0].set_ylabel("Densidad de Probabilidad", fontsize=12)
axis[0].grid(True)

axis[1].plot(data, cdf, "g-")
axis[1].set_xlabel("Valor de la Variable Aleatoria", fontsize=12)
axis[1].set_ylabel("Probabilidad Acumulada", fontsize=12)
axis[1].grid(True)

axis[2].plot(data, ppf, "b-")
axis[2].set_xlabel("Percentil", fontsize=12)
axis[2].set_ylabel("Valor de la Variable Aleatoria", fontsize=12)
axis[2].grid(True)

# - Visualización:

plt.tight_layout()
plt.show()
```

## Distribución Uniforme:

```python
# - Importe de las librerías a emplear:

from scipy.stats import uniform

# - Definición de todas las funciones relacionadas a la Distribución Uniforme:

def dunif(x, low = 0, high = 1):
    
    # - Calcula la estimación puntual de la distribución uniforme.
    
    result = uniform.pdf(x, loc = low, scale = (high - low))
    return result

def punif(q, low = 0, high = 1):
    
    # - Calcula el acumulado de la distribución uniforme.
    
    result = uniform.cdf(q, loc = low, scale = (high - low))
    return result

def qunif(p, low = 0, high = 1):
    
    # - Calcula la función cuantil a partir de la distribución uniforme.
    
    result = uniform.ppf(p, loc = low, scale = (high - low))
    return result

def runif(n, low = 0, high = 1):
    
    # - Genera variables aleatorias a partir de la distribución uniforme.
    
    result = uniform.rvs(loc = low, scale = (high - low), size = n)
    return result
```

```python
# - Los datos a emplear para la distribución uniforme son: Mínimo siendo loc=0, Máximo siendo scale=1.

np.random.seed(42)

print(f"Probabilidad de que una distribución uniforme (0, 1) tome el valor 0.5: {round(dunif(0.5, low = 0, high = 1) * 100, 2)}%")
print()
print(f"Probabilidad de que una distribución uniforme (0, 1) tome un valor menor que 0.5: {round(punif(0.5, low = 0, high = 1) * 100, 2)}%")
print()
print(f"Valor que de una distribución uniforme (0, 1) tiene una probabilidad acumulada de 0.9: {round(qunif(0.9, low = 0, high = 1), 2)}")
print()
print(f"Generación de 2000 valores aleatorios de una distribución uniforme (0, 1): {runif(2000, low = 0, high = 1)}")
```

```python
np.random.seed(42)

# - Declaración de variables:

low = 0
high = 1
data = np.arange(-5, 5, 0.01)

pmf = dunif(data, low = low, high = high)
cdf = punif(data, low = low, high = high)
ppf = qunif(data, low = low, high = high)

# - Distribución de los gráficos:

fig, axis = plt.subplots(1, 3, figsize = (20, 5))

# - Parámetros de los gráficos:

axis[0].plot(data, pmf, "r-")
axis[0].set_xlabel("Valor de la Variable Aleatoria", fontsize=12)
axis[0].set_ylabel("Densidad de Probabilidad", fontsize=12)
axis[0].grid(True)

axis[1].plot(data, cdf, "g-")
axis[1].set_xlabel("Valor de la Variable Aleatoria", fontsize=12)
axis[1].set_ylabel("Probabilidad Acumulada", fontsize=12)
axis[1].grid(True)

axis[2].plot(data, ppf, "b-")
axis[2].set_xlabel("Percentil", fontsize=12)
axis[2].set_ylabel("Valor de la Variable Aleatoria", fontsize=12)
axis[2].grid(True)

# - Visualización:

plt.tight_layout()
plt.show()
```
